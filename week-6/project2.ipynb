```python
import tkinter as tk

class Shipping:
    def __init__(self, destination, weight):
        self.destination = destination.lower()
        self.weight = weight

    def get_price(self):
        if self.destination == 'ibeju-lekki':
            return 2000 if self.weight < 10 else 5000
        elif self.destination == 'epe':
            return 4000 if self.weight < 10 else 10000
        else:
            return None

# Button logic
def compute():
    dest = dest_entry.get()
    
    try:
        weight = float(weight_entry.get())
    except ValueError:
        output_label.config(text="Enter a valid number for weight.")
        return

    if weight <= 0:
        output_label.config(text="Weight must be greater than 0.")
        return

    order = Shipping(dest, weight)
    price = order.get_price()

    if price is None:
        output_label.config(text="Location not supported (Epe or Ibeju only).")
    else:
        output_label.config(text=f"Price: ₦{price}")

# GUI setup
app = tk.Tk()
app.title("Shipping Cost Checker")
app.geometry("400x300")

tk.Label(app, text="Destination (Epe/Ibeju):").pack(pady=10)
dest_entry = tk.Entry(app, width=25)
dest_entry.pack(pady=5)

tk.Label(app, text="Weight (kg):").pack(pady=10)
weight_entry = tk.Entry(app, width=25)
weight_entry.pack(pady=5)

tk.Button(app, text="Check Price", command=compute).pack(pady=20)

output_label = tk.Label(app, text="")
output_label.pack(pady=10)

app.mainloop()
```
